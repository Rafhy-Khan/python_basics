## **Match-case**

Match-case is a control statement in Python that checks one value against multiple possible patterns and runs the matching block of code.It is Pythonâ€™s version of a switch statement, but more powerful because it supports pattern matching.

**Examples:**

```python
# Program: Interpret AI emotional state based on input code

emotion_code = 2

match emotion_code:
    case 1:
        print("AI is in learning mode")
    case 2:
        print("AI is curious and exploring new ideas")
    case 3:
        print("AI is optimizing performance")
    case 4:
        print("AI is debugging reality")
    case _:
        print("Unknown AI state detected")
```

```python
# Program: Decide spacecraft action based on signal received

signal = "LAND"

match signal:
    case "LAUNCH":
        print("Igniting boosters")
    case "ORBIT":
        print("Stabilizing orbit")
    case "LAND":
        print("Deploying landing gear")
    case "ABORT":
        print("Mission aborted safely")
    case _:
        print("Unknown command from mission control")
```

## **Difference between if-elif and match-case:**

1:if-elif checks conditions and match-case checks patterns

2:if can evaluate:

Comparisons (>, <)

Logical operators (and, or)

Complex expressions

3:match compares one value to specific patterns.

**Examples:**

```python
# Program: Analyze battery percentage using if-elif

battery = 75

if battery > 80:
    print("Battery Full")
elif battery > 50:
    print("Battery Moderate")
elif battery > 20:
    print("Battery Low")
else:
    print("Battery Critical")
```

```python
# Program: Analyze exact battery modes using match-case

battery_mode = "ECO"

match battery_mode:
    case "PERFORMANCE":
        print("High performance mode activated")
    case "BALANCED":
        print("Balanced energy usage")
    case "ECO":
        print("Energy saving mode enabled")
    case _:
        print("Unknown battery mode")
```

## **Nested-if:**

A nested if is an if statement inside another if statement.It is used when a second condition should only be checked if the first condition is true.

**Examples:**

```python
# Program: Check astronaut readiness

oxygen_level = 85
helmet_on = True

if oxygen_level > 70:
    if helmet_on:
        print("Astronaut ready for spacewalk")
    else:
        print("Helmet missing! Cannot proceed")
else:
    print("Oxygen level too low")
```

```python
# Program: Two-level security check

password_correct = True
otp_verified = True

if password_correct:
    if otp_verified:
        print("Access granted")
    else:
        print("OTP verification failed")
else:
    print("Incorrect password")
```

```python
# Program: Smart temperature control

temperature = 30
humidity = 70

if temperature > 25:
    if humidity > 60:
        print("Turn on AC and dehumidifier")
    else:
        print("Turn on AC only")
else:
    print("Temperature is comfortable")
```

```python
# Program: AI response moderation

content_safe = True
contains_sensitive_topic = True

if content_safe:
    if contains_sensitive_topic:
        print("Respond carefully with context awareness")
    else:
        print("Respond normally")
else:
    print("Content blocked")
```
